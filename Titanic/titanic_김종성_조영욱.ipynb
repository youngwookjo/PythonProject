```python
t_data_feature_f = t_data_feature.iloc[:0, :]
t_data_feature_l = t_data_feature.iloc[:0, :]
```

```python
d_rows = []
rows = []
for x, i in zip(t_data_feature["Age"].isnull(), range(len(t_data_feature["Age"]))):
    if x == True:
        d_rows.append(i)
    else:
        rows.append(i)
```

```python
for x, i in zip(d_rows, range(len(d_rows))):
    t_data_feature_f.ix[i] = t_data_feature.ix[x]
t_data_feature_f = t_data_feature_f.drop(columns=["Name", "Ticket"])
t_data_feature_f
```

```python
for x, i in zip(rows, range(len(rows))):
    t_data_feature_l.ix[i] = t_data_feature.ix[x]
t_data_feature_l = t_data_feature_l.drop(columns=["Name", "Ticket"])
t_data_feature_l
```

```python
from sklearn import svm, metrics
from sklearn.ensemble import RandomForestClassifier
# from sklearn.preprocessing import LabelEncoder
# le = LabelEncoder()

label = t_data_feature_l["Age"]
feature = t_data_feature_l.drop(columns=["Age"])
feature["Sex"] = t_data_feature_l.drop(columns=["Age"])["Sex"].apply(lambda x : 0 if x == "male" else 1)
feature["Embarked"] = feature["Embarked"].apply(lambda x : 0 if x == "S" else (1 if x == "Q" else 2))

test_feature = t_data_feature_f.drop(columns=["Age"])
test_feature["Sex"] = t_data_feature_l.drop(columns=["Age"])["Sex"].apply(lambda x : 0 if x == "male" else 1)
test_feature["Embarked"] = feature["Embarked"].apply(lambda x : 0 if x == "S" else (1 if x == "Q" else 2))

clf = RandomForestClassifier()
# clf = svm.SVC()
clf.fit(feature.values, label.values.astype(int))

pre = clf.predict(test_feature)
print(pre)
```
